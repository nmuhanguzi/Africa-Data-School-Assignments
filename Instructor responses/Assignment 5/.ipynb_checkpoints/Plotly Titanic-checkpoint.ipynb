# Loading Dataset

```python
import pandas as pd 
df = pd.read_csv("Titanic Data.csv")
df.head()
```

# Histogram

```python
df['Age'].head()
```

```python
df.iplot(kind="histogram", bins=20, theme="white", title="Passenger's Ages",xTitle='Ages', yTitle='Count', x = 'Age')
```

```python
px.histogram(df, x = 'Age', nbins = 20, title="Passenger's Ages", opacity = 0.7)
```

```python
df["male_age"]=df[df["Sex"]=="male"]["Age"]

df["female_age"]=df[df["Sex"]=="female"]["Age"]
```

```python
df[["male_age", "female_age"]]
```

```python
df[["male_age","female_age"]].iplot(kind="histogram", bins=20, theme="white", title="Passenger's Ages",
         xTitle='Ages', yTitle='Count')
```

```python
px.histogram(df, x = 'Age', nbins = 20, title="Passenger's Ages", color = 'Sex', opacity = 0.75)
```

```python
df.corr()
```

```python
px.imshow(df.corr(), title="Feature Correlation Matrix")
```

```python
df.corr().iplot(kind='heatmap', colorscale="Greens", title="Feature Correlation Matrix")
```

```python
#we will get help from pivot tables to get Fare values in different columns for each class.
df[['Pclass', 'Fare']].pivot(columns='Pclass', values='Fare').iplot(kind='box')
```

```python
px.box(df, x ='Pclass', y='Fare')
```

```python
survived_sex = df[df['Survived']==1]['Sex'].value_counts()
dead_sex = df[df['Survived']==0]['Sex'].value_counts()
df1 = pd.DataFrame([survived_sex,dead_sex])
df1.index = ['Survived','Dead']

px.bar(df1, barmode='stack', title='Survival by the Sex', opacity = 0.75)
```

```python
df1.iplot(kind='bar',barmode='stack', title='Survival by the Sex')
```

```python
df.iplot(kind="scatter", theme="white", x="Age", y="Fare", categories="Sex")
```

```python
px.scatter(df, x="Age",y="Fare", color="Sex")
```

```python
#converting Survived column to float64 to be able to use in plotly
df[['Survived']] = df[['Survived']].astype('float64', copy=False)
df.iplot(kind='bubble', x="Fare", y="Age", categories="Survived", size='Pclass', text='Name', xTitle='Fare', yTitle='Age')
```

```python
px.scatter(df, x="Fare", y="Age", color="Survived", size='Pclass', hover_name='Name')
```
